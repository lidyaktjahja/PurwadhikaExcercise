data : tips 

target : tip 

preproces v1 : 

1. One hot encoding : sex, smoker, time 

2. ordinal encoding : day 

3. no treatment : size, total_bill

data splitting 70 : 30 

model Tree max depth 3

```python
tips
```

```python
# ONE HOT ENCODING

tips['sex'].value_counts()
tips['sex'] = np.where(tips['sex']=='Male',1,0)

tips['smoker'].value_counts()
tips['smoker'] = np.where(tips['smoker']=='Yes',1,0)

tips['time'].value_counts()
tips['time'] = np.where(tips['time']=='Dinner',1,0)
```

```python
# ORDINAL ENCODING : day
tips['day'].value_counts()
tips['day'] = np.where(tips['day']=='Thur',4,(np.where(tips['day']=='Fri',5,np.where(tips['day']=='Sat',6,np.where(tips['day']=='Sun',7,0)))))
```

```python
tips.columns
```

```python
X = tips[['total_bill', 'sex', 'smoker', 'day', 'time', 'size']]
y = tips['tip']
```

```python
# X_robust = robust.fit_transform(X)
# X_robust = pd.DataFrame(X_robust, columns = ['Tip', 'Total_Bil'])
# X_robust
```

```python
tips
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    train_size=0.7,
    random_state=10
)
```

# RIDGE REGRESSION

```python
scaler = RobustScaler()
scaler.fit(X_train)
X_Train_Scaled = scaler.transform(X_train)
X_Test_Scaled = scaler.transform(X_test)
```

```python
y_train
```

```python
alpha = [100000, 10000, 1000, 100, 10, 1, 0.1, 0.01, 0.001]
testing_mse = []
training_mse = []
score = 0

for i in alpha :
    ridgereg = Ridge (alpha = i)
    ridgereg.fit(X_train, y_train)

    y_predict_train = ridgereg.predict(X_train)
    training_mse.append(mean_squared_error(y_train, y_predict_train))

    y_predict_test = ridgereg.predict(X_test)
    mse_score = mean_squared_error(y_test, y_predict_test)
    testing_mse.append(mse_score)

    if score > -mse_score :
        score = mse_score
        best_alpha = i
```

```python
length = range(len(alpha))

plt.figure(figsize=(15,10))
sns.lineplot(length, testing_mse)
sns.lineplot(length, training_mse)
plt.legend(['Testing Error','Training Error'])
plt.xticks(length,alpha)
plt.show()
```

```python
print('best_alpha : ',best_alpha)
print("MSE : ",score)
```

```python
ridge = Ridge(alpha = 0.001).fit(X_train,y_train)
print("Training set score : {:.2f}".format(ridge.score(X_train, y_train)))
print("Test set score : {:.2f}".format(ridge.score(X_test,y_test)))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
scaler = RobustScaler()
scaler.fit(X_train)
X_Train_Scaled = scaler.transform(X_train)
X_Test_Scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_Train_Scaled,y_train)
y_predict_class = knn.predict(X_test)
y_predict_class
print(confusion_matrix(y_test, y_predict_class))
print(classification_report(y_test,y_predict_class))
print('Akurasi Model Adalah : ', round(accuracy_score(y_test,y_predict_class)*100,2))
```

```python
k = range(1,100,2)
testing_accuracies = []
training_accuracies = []
```

```python
score = 0

for i in k:
  knn = KNeighborsClassifier(n_neighbors = i)
  knn.fit(X_Train_Scaled,y_train)

  y_predict_train = knn.predict(X_Train_Scaled)
  training_accuracies.append(accuracy_score(y_train,y_predict_train))

  y_predict_test = knn.predict(X_Test_Scaled)
  acc_score = accuracy_score(y_test,y_predict_test)
  testing_accuracies.append(acc_score)

  if score < acc_score:
    score = acc_score
    best_k = i
```

```python
plt.plot(training_accuracies,'b')
plt.plot(testing_accuracies,'orange')
```

```python
print("best K : ",best_k)
print('Akurasi : ',round(score*100,2))
```
