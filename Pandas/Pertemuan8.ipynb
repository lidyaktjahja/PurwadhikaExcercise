```python
def car_group(price,origin):
    if price == 0 :
        return 'No Price Data'
    elif origin == '0' :
        return 'No Origin Data'
    else :
        if price < 15 :
            return f'Cheap {origin}'
        elif price < 30 :
            return f'Moderate {origin}'
        else :
            return f'Expensive {origin}'
```

```python
data['Price'].fillna(0,inplace = True)
data['Origin'].fillna('0', inplace = True)
```

```python
data['Price.Origin'] = data.apply(lambda x : car_group(x['Price'], x['Origin']), axis = 1)
data[data.Price == 0][['Price','Origin','Price.Origin']]
```
